```python
import matplotlib.pyplot as plt

# Example data (replace this with your actual data)
time_values = [0, 4.00, 18.00, 28.00, 46.00, 50.00, 68.00, 78.00, 82.00]
cpu_utilization_values = [0, 4.35,  2.17, 10.87,  2.17,  4.35,  2.17, 10.87,  2.17]

# Plotting the data
plt.plot(time_values, cpu_utilization_values, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Time')
plt.ylabel('CPU Utilization (%)')
plt.title('Time vs CPU Utilization')

# Display the plot
plt.show()
```

```python
import datetime

# Sample data (replace with your actual data)
time_points = [
    "2024-02-08 10:00:00",
    "2024-02-08 11:00:00",
    "2024-02-08 12:00:00",
    # Add more time points as needed
]
utilization_values = [70, 80, 60]  # Replace with your actual utilization values

# Convert time strings to datetime objects for better plotting
time_points = [datetime.datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S") for time_str in time_points]

# Plotting
plt.plot(time_points, utilization_values, marker='o')
plt.title('Resource Utilization Over Time')
plt.xlabel('Time')
plt.ylabel('Utilization (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
timestamps =  [0, 4.00, 18.00, 28.00, 46.00, 50.00, 68.00, 78.00, 82.00, 92.00]
utilization = [0, 0, 4.35,  2.17, 10.87,  2.17,  4.35,  2.17, 10.87,  2.17]

# Plotting the resource utilization over time
plt.step(timestamps, utilization, where='pre')
plt.xlabel('Time (ms)')
plt.xticks(timestamps)
plt.ylabel('Resource Utilization (%)')
plt.title('Resource Utilization Over Time')
plt.show()
```
